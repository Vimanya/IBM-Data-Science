## 1.Explore Dataset

tools & library used : web scraping using BeautifulSoup package

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page =requests.get(url)
soup = BeautifulSoup(page.text,'html.parser')
table_1= soup.find('table',{"class":"wikitable"})
print(table_1)
```

to get only texts in table

```python
labels = []
for postal_code in table_1.find_all('tbody'):
    rows =  postal_code.find_all('tr')
    for row in rows:
        details = row.find_all('td')
        for detail in details:
            r =detail.text.strip()
            labels.append(r)
            print(r)
```

## 2.Create Dataframe

Divide the data into different dataframes so I took 3 dataframes to assign values.

```python
df = pd.DataFrame(labels)
df1 = pd.DataFrame(df.iloc[::3])
df2 = pd.DataFrame(df.iloc[1::3])
df3 = pd.DataFrame(df.iloc[2::3])
```

```python
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
df3 = df3.reset_index(drop=True)
```

```python
df1 = df1.rename(columns={0: "Postal Code"})
df2 = df2.rename(columns={0:'Borough'})
df3=df3.rename(columns={0:'Neighbourhood'})
```

```python
df_Tor = pd.concat([df1,df2,df3],axis =1)
df_Tor
```

## 3.Clean the data

Ignore cells with a borough that is 'Not assigned'.

```python
df_Tor.drop(df_Tor.loc[df_Tor['Borough']=='Not assigned'].index, inplace=True)
df_Tor = df_Tor.reset_index(drop=True)
df_Tor
```

```python
df_Tor['Neighbourhood'] = df_Tor.Neighbourhood.str.split('/')
df_Tor.head()
```

If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
for index,row in df_Tor.iterrows():
    if df_Tor['Neighbourhood'].empty:
        df_Tor['Neighbourhood'] = df_Tor['Borough']
```
